```python
import pandas as pd
import matplotlib

arquivo = pd.read_csv("bla/votacao_secao_2022_BR.csv", sep=";", encoding="iso-8859-1")
arquivo
```

```python
data = pd.DataFrame(arquivo, columns = ["NR_ZONA", "NR_SECAO", "NR_PARTIDO", "QT_APTOS", "QT_VOTOS"], dtype=int) \
  .assign(porcentagem = lambda x: x.QT_VOTOS / x.QT_APTOS * 100)

is_lula = data["NR_PARTIDO"] == 13
is_bolsonaro = data["NR_PARTIDO"] == 22

lula = data.where(is_lula).dropna()
bolsonaro = data.where(is_bolsonaro).dropna()

lula, bolsonaro
```

```python
bolsonaro[["porcentagem"]].hist(bins=100)
lula[["porcentagem"]].hist(bins=100)
```
